```python
import pandas as pd
import re
import tarfile
```

```python
file_path = "../UAS/WhatsApp Chat with courteous.txt"

data = []
with open(file_path, 'r', encoding='utf-8') as file:
    for line in file:
        match = re.match(r"(\d+/\d+/\d+), (\d+:\d+\s[AP]M) - (.*?): (.*)", line.strip())
        if match:
            tanggal, waktu, nama, pesan = match.groups()
            data.append({"Tanggal": tanggal, "Waktu": waktu, "Nama Kontak": nama, "Pesan": pesan})
```
